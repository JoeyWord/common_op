```python
import re
trs=table_need.find_all('tr')
head=[]
contents=[]
count=0
for tr in trs:
    count+=1
    if count==1:
        tds=tr.find_all('td')
        for td in tds:
            text=td.get_text()
            text=re.sub('\s+','',text)
            head.append(text)
    else:
        tds=tr.find_all('td')
        tmp=[]
        for td in tds:
            text=td.get_text()
            text=re.sub('\s+','',text)
            tmp.append(text)
        contents.append(tmp)
print("head info:\n",head)
print("content info:\n",contents)
```
